```python

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 2: Prepare data
hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
marks = [20, 25, 35, 45, 50, 60, 70, 80, 85]

df = pd.DataFrame({'Hours': hours, 'Marks': marks})
print(df)

# Step 3: Visualize data
plt.scatter(df['Hours'], df['Marks'], color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.title("Study Hours vs Marks")
plt.show()

# Step 4: Split into train/test
X = df[['Hours']]
y = df['Marks']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 5: Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 6: Predict
predicted = model.predict(X_test)
print("Predicted marks:", predicted)

# Step 7: Test new value
hours_studied = [[7.5]]
predicted_marks = model.predict(hours_studied)
print(f"If you study {hours_studied[0][0]} hours, predicted marks = {predicted_marks[0]:.2f}")
```
